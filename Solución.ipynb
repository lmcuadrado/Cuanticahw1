## Solución para la ecuación diferencial del oscilador armónico

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defino el rango de tiempo sobre el que voy a hallar las posiciones
min_t = 0.0
max_t = 20.0
dt=0.01
n= int((max_t - min_t)/dt) #cantidad de puntos o datos que se tomaran

#Creo los arreglos para el tiempo, posición y velocidad
t = np.zeros(n)
x = np.zeros(n)
v = np.zeros(n)

#Defino los valores iniciales

w=1   #constante/frecuencia
t[0] = min_t
x[0] = 2.0
v[0] = 0.0
```

```python
#a=dv/dt=-wx
def a(t,x):          
    return -w*x

#Solución analitica de la ecuación por el metodo de euler
for i in range(0,n-1):    
    t[i+1] = t[i] + dt         #Avance en el tiempo
    v[i+1] = v[i] + a(t[i],x[i])*dt  #Resuelvo dv/dt =-wx --> dv=-w*x*dt
    x[i+1] = x[i] + dt * v[i]  #Resuelvo dx/dt=v --> dx=v*dt
```

```python
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```
